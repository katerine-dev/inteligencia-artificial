# Groupby
O método groypby permite agrudar linhas de dados e chamar funções de agregação.

```python
import numpy as np
import pandas as pd
```

```python
data = {'Company': ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
        'Person': ['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
        'Sales': [200, 120, 340, 124, 243, 350]}
```

```python
df = pd.DataFrame(data)
df
```

Agora você pode utilizar o método groupby() para agrupar linhas com base em um nome de coluna. Por exemplo, vamos agrupar com base na empresa. Isso criará um objeto DataFrameGroupBy

```python
df.groupby('Company')
```

Você pode salvar esse objeto como uma nova váriavel

```python
by_comp = df.groupby('Company')
```

E então chamar métodos agregadores de objetos

```python
by_comp.mean()
```

```python
df.groupby('Company').mean()
```

Mais alguns exemplos de métodos de agregação

```python
by_comp.std()
```

```python
by_comp.min()
```

```python
by_comp.max()
```

```python
by_comp.count()
```

```python
by_comp.describe().transpose()
```

```python
by_comp.describe().transpose()['GOOG']
```

## Mesclando, Unindo e Concatenand

Existem 3 maneiras principais de combinar DataFrames: Mesclando, Unindo e Concatenando

### Concatenação
A concatenação basicamente une os DataFrames. Tenha em mente que as dimensÕes devem corresponder ao longo do eixo em que você está concatenando. Você pode usar *pd.concat()* e passar uma lista de DataFrames pra concatenar

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                    'B':['B4','B5','B6','B7'],
                    'C':['C4','C5','C6','C7'],
                    'D':['D4','D5','D6','D7']})

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                    'B':['B8','B9','B10','B11'],
                    'C':['C8','C9','C10','C11'],
                    'D':['D8','D9','D10','D11']})
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
pd.concat([df1,df2,df3], axis=1) # axis = 1 ele coloca um do lado do outro
```

### Mesclando
A função *merge()* permite mesclar DataFrames usando uma lógica semeslhante a mesclagem de tabelas SQL

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right, how='inner', on = 'key')
```

Um exemplo mais complicado:

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                      'A':['A0','A1','A2','A3'],
                      'B':['B0','B1','B2','B3'],})

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                     'key2': ['K0', 'K0', 'K0', 'K0'],
                      'C':['C0','C1','C2','C3'],
                      'D':['D0','D1','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on =['key1', 'key2'])
```

```python
pd.merge(left, right,how = 'outer', on =['key1', 'key2'])
```

```python
pd.merge(left, right,how = 'right', on =['key1', 'key2'])
```

```python
pd.merge(left, right,how = 'left', on =['key1', 'key2'])
```

### Junção
A junção é um método conveniente para combinar as colunas de dois DataFrames indexados de forma potencialmente diferente em um único DataFrame resultante

```python
left = pd.DataFrame({'A':['A0','A1','A2','A3','A4','A5'],
                     'B':['B0','B1','B2','B3','B4','B5'],
                     'index':['K0','K1','K2','K3','K4','K5']})

right = pd.DataFrame({'C':['C0','C1','C2'],
                      'D':['D0','D1','D2'],
                      'index': ['K0', 'K2', 'K3']})
```

```python
left
```

```python
right
```

```python
left.join(right, lsuffix = '_left', rsuffix = '_right')
```
